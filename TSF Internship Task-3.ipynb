### Data Visualization

```python
# correlation analysis using heatmap

df.corr(numeric_only= True)
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only= True), annot=True)
plt.show()
```

```python
# visualizing using the scatter plot

num = ['Sales','Quantity','Discount','Profit']
num_df = df[num]
g = sns.PairGrid(num_df)
g.map(sns.scatterplot)
```

- Profit and Sales - The profit is clustered around the 0 and no visible increase in profits as sale increases. It shows that the sales does not affect the profit.
- Qquantity is also not correlated to profit. As discount increases, we can see that profit decreases.

#### Bivariate Analysis

```python
categorical_count= df.groupby("Category").sum(numeric_only= True)
print(categorical_count)
```

```python
categorical_count[['Sales', 'Profit']].plot.bar()
```

```python
subcategorical_sum= df.groupby("Sub-Category").sum(numeric_only=True)
subcategorical_sum
```

```python
subcategorical_sum[['Sales', 'Profit']].plot(kind = 'bar', figsize = [20,8])
plt.ylabel('Total Sales Vs Profits',fontsize = 18)
```

```python
# category vs sub-category analysis

plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount',y='Profit',data=df)
plt.title("Discount vs Profit")
```
